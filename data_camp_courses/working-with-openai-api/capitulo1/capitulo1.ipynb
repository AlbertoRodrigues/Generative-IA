## Uma API é um serviço dispobilizado onde podemos fazer requisições para uma determinada tarefa e receber a resposta



## Na API da OpenAI nós enviamos vários parâmetros como: modelo, input e roles


## Há vários endpoints da API, podemos usar o ChatCompletion, o Completion, entre outros


## Primeira requisição



```python
# Import openai
import openai
```

```python
response = openai.Completion.create(
  # Specify the correct model
  model="gpt-3.5-turbo-instruct",
  prompt="Who developed ChatGPT?"
)

print(response)
```

```python
# Resposta do chatgpt
print(response["choices"][0]["text"])
```
